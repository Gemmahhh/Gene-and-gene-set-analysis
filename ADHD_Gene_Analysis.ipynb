```python
# viewing the first  few records in the annotation file in a more structured manner 


file_path = '/Users/cokpe/Documents/MAGMA_project/annotation_results.genes.annot'

# As seen from the  output above, the first 2 lines are comments so we  are going to skip them 
with open(file_path, 'r') as file:
    for _ in range(2):
        next(file)

    
    data = []
    for line in file:
        parts = line.strip().split()
        identifier = parts[0]
        chrom_loc = parts[1]
        snps = parts[2:]
        chrom, start, end = chrom_loc.split(':')
        data.append([identifier, chrom, start, end] + snps)

# the line of code below determines the maximum number of SNPs in any row
max_snps = max(len(row) - 4 for row in data)


for row in data:
    while len(row) - 4 < max_snps:
        row.append(None)

#list for the column names definition 
columns = ['Identifier', 'Chromosome', 'Start', 'End'] + [f'SNP_{i+1}' for i in range(max_snps)]

# Create DataFrame
df = pd.DataFrame(data, columns=columns)

# Display the first few rows
print(df.head())

```

```python
# now, lets clean up the gene set data 

input_file_path = '/Users/cokpe/Documents/MAGMA_project/gene_set_folder/c2.cp.v2023.2.Hs.entrez.gmt'
output_file_path = '/Users/cokpe/Documents/MAGMA_project/gene_set_folder/gene_set_data.txt'


with open(input_file_path, 'r') as file:
    lines = file.readlines()

processed_data = []

for line in lines:
    parts = line.strip().split('\t')
    gene_set_name = parts[0]
    gene_ids = parts[2:]  # Skip the URL part
    processed_line = gene_set_name + '\t' + '\t'.join(gene_ids)
    processed_data.append(processed_line)

with open(output_file_path, 'w') as f:
    for line in processed_data:
        f.write(line + '\n')

for line in processed_data[:5]: 
    print(line)

```

```python
# Define the path to the processed output file
processed_file_path = '/Users/cokpe/Documents/MAGMA_project/gene_set_folder/gene_set_data.txt'
unique_gene_sets_file_path = '/Users/cokpe/Documents/MAGMA_project/gene_set_folder/unique_gene_sets.txt'

unique_gene_sets = set()

with open(processed_file_path, 'r') as f:
    for line in f:
        gene_set_name = line.split('\t')[0] 
        unique_gene_sets.add(gene_set_name)

with open(unique_gene_sets_file_path, 'w') as f:
    for gene_set_name in sorted(unique_gene_sets): 
        f.write(gene_set_name + '\n')

print(f"Unique gene set names have been saved to {unique_gene_sets_file_path}")
```

# INTERPRETING THE MAGMA GENE  SET ANALYSIS RESULT

```python
# first, lets view the result

results_file = '/Users/cokpe/Documents/MAGMA_project/gene_set_analysis_result.gsa.out'
results_df = pd.read_csv(results_file, sep='\t')
print(results_df.head(20))
```

Now, lets put the dataset in the right format

```python
import pandas as pd

# Load the file without headers to inspect the structure
results_file = 'gene_set_analysis_result.gsa.out'
set_results = pd.read_csv(results_file, delim_whitespace=True, comment='#')

# Display the DataFrame to inspect columns and data
print(set_results.head())
print(set_results.columns)
```

```python
#  lets check for significant gene sets 

significance_threshold = 0.05
significant_gene_sets = set_results[set_results['P'] < significance_threshold]
print(significant_gene_sets)
```

```python
# the bar plot below is used to show the top 10 significant gene sets
top_gene_sets = significant_gene_sets.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_gene_sets['VARIABLE'], -np.log10(top_gene_sets['P']))
plt.xlabel('-log10(P-value)')
plt.title('Top 10 Significant Gene Sets')
plt.gca().invert_yaxis()
plt.show()
```

Now, lets plot MAGMA beta values against the negative logarithm (base 10) of Bonferroni-corrected p-values

```python
# creating and populating the neg log column 

set_results['neg_log_p'] = -np.log10(set_results['P'])
```

```python
print(set_results.head())
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(set_results['BETA'], set_results['neg_log_p'], color='green', alpha=0.7, s=50)
plt.xlabel('Beta Value', fontsize=12)
plt.ylabel('-log10(Bonferroni p-value)', fontsize=12)
plt.title('MAGMA Beta vs -log10(Bonferroni p-value)', fontsize=14)
plt.grid(True)
plt.show()
```

The is a scatter plot of MAGMA beta values against the negative logarithm (base 10) of Bonferroni-corrected p-values for each gene set. Each point represents a gene set, with the x-axis indicating the effect size (beta value) and the y-axis representing the significance of the association. 


As seen from the chart above, clusters of points concentrated around a beta value of 0.5 or below on the x-axis suggest gene sets with relatively low effect sizes in their association ADHD. The distribution indicates that these gene sets may not exert strong individual effects on the trait but could collectively contribute to its manifestation. 

Also, their presence could suggest the involvement of polygenic mechanisms or complex interactions among multiple genes within these sets. 

Further investigation into these gene sets may uncover subtle yet meaningful contributions to the trait's genetic architecture, and also reveal insights into its underlying biology or pathophysiology.

Now, lets plot MAGMA number of genes against the negative logarithm (base 10) of Bonferroni-corrected p-values

```python
plt.figure(figsize=(10, 8))
plt.scatter(set_results['NGENES'], set_results['neg_log_p'], color='green', alpha=0.7, s=50)
plt.xlabel('Number of Genes per Gene Set', fontsize=12)
plt.ylabel('-log10(Bonferroni p-value)', fontsize=12)
plt.title('MAGMA Number of Genes vs -log10(Bonferroni p-value)', fontsize=14)
plt.grid(True)
plt.show()
```
